```python
# Crear series de conteo de productos por categoría
df_tienda = tienda["Categoría del Producto"].value_counts()
df_tienda2 = tienda2["Categoría del Producto"].value_counts()
df_tienda3 = tienda3["Categoría del Producto"].value_counts()
df_tienda4 = tienda4["Categoría del Producto"].value_counts()

# Función para graficar
def graficar_ventas(nombre_tienda, datos_ventas, color):
    plt.figure(figsize=(6, 3))
    barras = plt.barh(datos_ventas.index, datos_ventas.values, color=color)
    plt.title(f'Ventas por Categoría {nombre_tienda}')
    plt.xlabel('Ventas Totales')
    plt.ylabel('Categoría')
    plt.grid(axis='x', linestyle='--', alpha=0.8)
    plt.gca().invert_yaxis()

    for barra in barras:
        valorx = barra.get_width()
        plt.text(valorx, barra.get_y() + barra.get_height()/2, f"{valorx:,.0f}", va='center', fontsize=8)

# Crear gráficos
graficar_ventas("Tienda 1", df_tienda, "blue")
graficar_ventas("Tienda 2", df_tienda2, "red")
graficar_ventas("Tienda 3", df_tienda3, "green")
graficar_ventas("Tienda 4", df_tienda4, "brown")

plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Función para mostrar los 5 productos más y menos vendidos
def mostrar_top_bottom_5(tienda_actual, nombre_de_tienda):#Creamos una función con los parámetros tienda_actual y nombre_de_tienda
    conteo = tienda_actual['Producto'].value_counts() # 'tienda['Producto']' selecciona la columna de productos, y 'value_counts()' cuenta cuántas veces aparece cada producto en esa columna, devolviendo un conteo de ventas por producto.

    top5 = conteo.head(5) #selecciona los 5 primeros productos con más ventas

    bottom5 = conteo.tail(5).sort_values(ascending=True)  #'tail(5)' selecciona los 5 últimos productos con menos ventas, y 'sort_values(ascending=True)' los ordena de menor a mayor

    print(f"{nombre_de_tienda}") # Imprime el nombre de la tienda
    print("\n5 productos más vendidos:")  # Imprime encabezado para los más vendidos
    for producto, ventas in top5.items(): # Itera sobre los 5 productos más vendidos y muestra su nombre y cantidad de ventas
        print(f"{producto}: {ventas} ventas")

    print("\n5 productos menos vendidos:") # Imprime encabezado para los menos vendidos
    for producto, ventas in bottom5.items():   # Itera sobre los 5 productos menos vendidos y muestra su nombre y cantidad de ventas
        print(f"{producto}: {ventas} ventas")

    print("-" * 50)

    return top5, bottom5 # Retorna los datos de los productos más vendidos y menos vendidos para usarlos después si es necesario

# Llamadas a la función para cada tienda y guardado de los resultados en variables
top1, bottom1 = mostrar_top_bottom_5(tienda, "Tienda 1")
top2, bottom2 = mostrar_top_bottom_5(tienda2, "Tienda 2")
top3, bottom3 = mostrar_top_bottom_5(tienda3, "Tienda 3")
top4, bottom4 = mostrar_top_bottom_5(tienda4, "Tienda 4")

#NUEVO
#value_counts(): Cuenta la frecuencia de cada valor único en una columna. Ej. Cuenta las veces que aparece cada producto en la columna 'Producto' de la tienda.
#head(): Devuelve los primeros 5 elementos de la serie.  Ej. Los 5 productos más vendidos.
#tail(): Devuelve los últimos 5 elementos de la serie. Ej. Los 5 productos menos vendidos.
#sort_values(ascending=True): Ordena los valores de manera ascendente (de menor a mayor). Ej. Se usa en los productos menos vendidos para asegurarse de que se muestren del que menos vendió al que más vendió.
#Si usas sort_values(ascending=False), los valores se ordenarán de manera descendente (de mayor a menor).
#items(): Permite recorrer una serie (como un diccionario), devolviendo tanto la clave (producto) como el valor (ventas) para poder imprimirlos o utilizarlos en un ciclo.
```

```python
# Función para graficar top y bottom 5 productos
def graficar_top_bottom(top, bottom, nombre_tienda):
    # Gráfico Top 5
    plt.figure(figsize=(8, 4))
    top.plot(kind='bar', color='skyblue')
    plt.title(f"Top 5 productos más vendidos - {nombre_tienda}")
    plt.xlabel("Producto")
    plt.ylabel("Cantidad de ventas")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Gráfico Bottom 5
    plt.figure(figsize=(8, 4))
    bottom.plot(kind='bar', color='salmon')
    plt.title(f"Top 5 productos menos vendidos - {nombre_tienda}")
    plt.xlabel("Producto")
    plt.ylabel("Cantidad de ventas")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Llamadas para cada tienda
graficar_top_bottom(top1, bottom1, "Tienda 1")
graficar_top_bottom(top2, bottom2, "Tienda 2")
graficar_top_bottom(top3, bottom3, "Tienda 3")
graficar_top_bottom(top4, bottom4, "Tienda 4")
```

# 5. Valor del envío promedio por tienda

```python
def costo_envio_promedio(tienda_actual, nombre_tienda): # Función para calcular el costo de envío promedio de cada tienda
    costo_promedio = tienda_actual['Costo de envío'].mean()  # Calcula el promedio de la columna 'Costo de envío'
    print(f"Promedio de costo de envío para {nombre_tienda}: {costo_promedio:.2f} ") #Accede a la columna "Costo de envío" de la tienda_actual y usa .mean() para calcular su promedio. El resultado se guarda en la variable costo_promedio.

# Llamar a la función para cada tienda
costo_envio_promedio(tienda, "Tienda 1")
costo_envio_promedio(tienda2, "Tienda 2")
costo_envio_promedio(tienda3, "Tienda 3")
costo_envio_promedio(tienda4, "Tienda 4")
```

```python
import matplotlib.pyplot as plt

# Calcular los promedios de costo de envío
promedios_envio = {
    "Tienda 1": tienda['Costo de envío'].mean(),
    "Tienda 2": tienda2['Costo de envío'].mean(),
    "Tienda 3": tienda3['Costo de envío'].mean(),
    "Tienda 4": tienda4['Costo de envío'].mean()
}

# Crear el gráfico de barras horizontal
plt.figure(figsize=(8, 5))
plt.barh(list(promedios_envio.keys()), list(promedios_envio.values()), color='mediumseagreen')
plt.xlabel("Costo de envío promedio")
plt.title("Costo de envío promedio por tienda")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
